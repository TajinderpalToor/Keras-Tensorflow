```python
import tensorflow as tf
```

```python
#initilization of tensors
#tensors are just n dimensional blocks of data, which are the basis of all machine learning and deep learning operations

x=tf.constant(4,shape=(1,1),dtype=tf.float32)
print(x)

y=tf.constant([[1,2,3],[4,5,6]])
print(y)

#can also use the zeros function below
z=tf.ones((3,3))
print(z)

#I for the identity matrix
a=tf.eye(3)
print(a)

#can also take stuff from distributions

x=tf.random.normal((3,3),mean=0,stddev=1)
print(x)

x=tf.random.uniform((1,3),minval=0,maxval=1)
print(x)

#tf.range function
# can use tf.cast to convert datatypes


#Mathematical Operations:

x=tf.constant([1,2,3])
y=tf.constant([9,8,7])
z=tf.add(x,y) # or z=x+y
print(z)
b=tf.subtract(x,y)
c=tf.multiply(x,y)
d=tf.tensordot(x,y,axes=1)
print(d)
x=tf.random.normal((2,3))
y=tf.random.normal((3,4))
result=tf.matmul(x,y)
print(result)

```

```python
#indexing

x=tf.constant([0,1,1,1,2,3,7,8])

#the below just means print all but skip by 2, or whatever number you put after the semicolons
print(x[::2])

#you want to see just certain indices
temp=tf.constant([0,4,5])
print(tf.gather(x,temp))
```
